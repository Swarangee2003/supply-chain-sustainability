```python
import openrouteservice
import pandas as pd
import time

#ORS API key
api_key = "eyJvcmciOiI1YjNjZTM1OTc4NTExMTAwMDFjZjYyNDgiLCJpZCI6ImMxMjhjZjA4ZGJkNTQwMWM5NmU1MGMwOWNkNGU2OTY2IiwiaCI6Im11cm11cjY0In0=" 
client = openrouteservice.Client(key=api_key)

# City coordinates (lat, lon)
city_coords = {
    "Delhi": [77.1025, 28.7041],
    "Mumbai": [72.8777, 19.0760],
    "Hyderabad": [78.4867, 17.3850],
    "Pune": [73.8567, 18.5204],
    "Visakhapatnam": [83.2185, 17.6868],
    "Kochi": [76.2673, 9.9312],
    "Patna": [85.1376, 25.5941],
    "Vadodara": [73.1812, 22.3072],
    "Sambhajinagar": [75.3231, 19.8762],
    "Agra": [78.0081, 27.1767],
    "Rajkot": [70.8022, 22.3039],
    "Ludhiana": [75.8573, 30.9000],
    "Dehradun": [78.0322, 30.3165],
    "Chandigarh": [76.7794, 30.7333],
    "Kanpur": [80.3319, 26.4499],
    "Lucknow": [80.9462, 26.8467],
    "Ahmedabad": [72.5714, 23.0225],
    "Jaipur": [75.7873, 26.9124],
    "Chennai": [80.2707, 13.0827],
    "Bangalore": [77.5946, 12.9716],
    "Indore": [75.8577, 22.7196],
    "Nagpur": [79.0882, 21.1458],
    "Shimla": [77.1734, 31.1048],
    "Gurugram": [77.0266, 28.4595],
    "Jamnagar": [70.0577, 22.4707],
    "Surat": [72.8311, 21.1702],
    "Mysore": [76.6394, 12.2958],
    "Tiruvananthapuram": [76.9366, 8.5241],
    "Bhopal": [77.4126, 23.2599]
}


fuel_efficiency_kmpl = 4  # km per litre
emission_factor = 2.68    # kg CO2 per litre
trips_per_month = 10

# Prepared all valid (unique) route combinations
cities = list(city_coords.keys())
routes = [(source, dest) for i, source in enumerate(cities) 
          for dest in cities[i+1:] if source != dest]

data = []

# Loop over each route
for i, (source, dest) in enumerate(routes, start=1):
    coords = [city_coords[source], city_coords[dest]]
    
    try:
        result = client.directions(coords, profile='driving-car')
        distance_km = result['routes'][0]['summary']['distance'] / 1000  # in km
        print(f"{i}: {source} → {dest} = {round(distance_km, 2)} km")

        # Calculations
        litres_used = distance_km / fuel_efficiency_kmpl
        co2_per_trip = litres_used * emission_factor
        co2_per_month = co2_per_trip * trips_per_month

        # Add record
        data.append({
            "Route_ID": f"R{i:03}",
            "Source": source,
            "Destination": dest,
            "Distance_km": round(distance_km, 2),
            "Fuel_Used_Litres": round(litres_used, 2),
            "CO2_kg_Per_Trip": round(co2_per_trip, 2),
            "Monthly_Trips": trips_per_month,
            "Total_Monthly_CO2_kg": round(co2_per_month, 2)
        })

        time.sleep(1)  # Respect rate limits 

    except Exception as e:
        print(f"Error with {source} to {dest}: {e}")
        continue

# Save to CSV
df = pd.DataFrame(data)
df.to_csv("actual_supply_routes.csv", index=False)
print("✅ File saved as 'actual_supply_routes.csv'")
```
